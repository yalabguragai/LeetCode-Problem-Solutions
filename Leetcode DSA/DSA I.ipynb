Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

*Test Cases*
#Test Case 1:
Input: nums = [1,2,3,1]  
Output: true

#Test Case 2:
Input: nums = [1,2,3,4]  
Output: false

This first solution is using a hashtable in the form of dictionary.Iterating through the loop,if we encounter same number, the value  will be increased. And at last the condition is checked if the number's key value is greater than 1 or not.

```python
def containsDuplicate(nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        count={}
        for i in nums:
          if i in count:
            count[i]+=1
            #print(count[i])
          else:
            count[i]=1
            #print(count[i])
        print(count)
        return max(count.values()) > 1
            
```

```python
containsDuplicate([1,2,3,1])
```

This another solution is  using set function. A set function will only keep one copy of the particular item.It means, if the number is repeated,its length decreases and if the number in the *nums* array is unique, it prints the total length of the array.

If the length of the set and length of array is same, we can assume False  contition  and if its same, we can say the condition is True.

```python
def containsDuplicate(nums):
  a=set(nums)
  print(a)
  print(len(a))
  print(len(nums))
  if len(a)==len(nums):
    return False
  return True
```

```python
containsDuplicate([1,2,3,4])
```

Another solution is using the sort function.The *nums.sort()*  function will sort the numbers in order. And we can iterate through the numbers to check the occurance of that number.

```python
def containsDuplicate(nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        nums.sort()
        for i in range(len(nums)-1):
          print(nums[i])
          print(nums[i+1])
          if nums[i]==nums[i+1]:
            print(True)
          else:
            print(False)
```
