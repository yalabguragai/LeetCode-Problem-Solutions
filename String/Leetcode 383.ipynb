**Ransomnote Problem**  
Leetcode 383

Given two strings ransomNote and magazine, return true if ransomNote can be constructed by using the letters from magazine and false otherwise.

Each letter in magazine can only be used once in ransomNote.

**Test Cases**  
*Case I*  
Input: ransomNote = "a", magazine = "b"  
Output: false

*Case II*  
Input: ransomNote = "aa", magazine = "aab"  
Output: true  

The first solution is using Brute Force Approach.
1. Initialize two dictionaries, each for storing the key pair values of the string ransomNote and magazine.
2. Iterate through each string and store the occurance of each character in ransomNote and magazine string in ransomNote_dict and magazine_dict respectively.
3. Check if the frequency of characters in ransomNOte_dict is present in magazine_dict
  3.1. If present, return True.
  3.2. If not present, return False

**Note**
This is a brute force method and not optimized solution.

```python
def canConstruct(ransomNote: str, magazine: str) -> bool:
  magazine_dict, ransomNote_dict = {}, {}
  for letter in magazine:
    if letter in magazine_dict:
      magazine_dict[letter]+=1
    else:
      magazine_dict[letter]=1
  print (magazine_dict)
    #same process for ransomnote
  for letter in ransomNote:
    if letter in ransomNote_dict:
      ransomNote_dict[letter]+=1
    else:
      ransomNote_dict[letter]=1
  print(ransomNote_dict)
    #check contitions now
  for letter in ransomNote_dict:
    if letter not in magazine_dict:
      return False
    if ransomNote_dict[letter]> magazine_dict[letter]:
      return False
  return True
```

This second solution is the optimied version of the previous solution. 
1. Create only one hash table of magazine string as magazine_dict and store the occurances. 
2. Iterate through each characters in ransomNote and check its avaiability in magazine_dict.  
  2.1. If not present, return False
  2.2. If present, decrease the key value of that particular character    from magazine_dict dictionary.
3. If passed through above all conditions, return True.

**Note**
This is the optimized solution having time complexity O(M+N) and space complexity O(M).  
95.57%

```python
def canConstructt(ransomNote, magazine):
  magazine_dict={}
  for ch in magazine:
    if ch in magazine_dict:
      magazine_dict[ch]+= 1
    else:
      magazine_dict[ch]=1
              #comparisn with RansomNote string
  for ltr in ransomNote:
    if ltr not in magazine_dict or magazine_dict[ltr]==0:
      return False
    else:
      magazine_dict[ltr] -=1
  return True
```
